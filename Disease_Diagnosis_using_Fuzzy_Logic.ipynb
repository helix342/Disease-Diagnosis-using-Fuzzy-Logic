```python
import numpy as np
import pandas as pd
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
df = pd.read_csv("/content/Testing.csv")
df.head()
```

```python
itching = ctrl.Antecedent(np.arange(0, 1.1, 0.1), 'itching')
skin_rash = ctrl.Antecedent(np.arange(0, 1.1, 0.1), 'skin_rash')
shivering = ctrl.Antecedent(np.arange(0, 1.1, 0.1), 'shivering')
nodal_skin_eruptions = ctrl.Antecedent(np.arange(0, 1.1, 0.1), 'nodal_skin_eruptions')
continuous_sneezing = ctrl.Antecedent(np.arange(0, 1.1, 0.1), 'continuous_sneezing')
chills = ctrl.Antecedent(np.arange(0, 1.1, 0.1), 'chills')
joint_pain = ctrl.Antecedent(np.arange(0, 1.1, 0.1), 'joint_pain')
stomach_pain = ctrl.Antecedent(np.arange(0, 1.1, 0.1), 'stomach_pain')
acidity = ctrl.Antecedent(np.arange(0, 1.1, 0.1), 'acidity')
ulcers_on_tongue = ctrl.Antecedent(np.arange(0, 1.1, 0.1), 'ulcers_on_tongue')

Fungal_infection = ctrl.Consequent(np.arange(0, 1.1, 0.1), 'Fungal infection')
Allergy = ctrl.Consequent(np.arange(0, 1.1, 0.1), 'Allergy')
Drug_Reaction = ctrl.Consequent(np.arange(0, 1.1, 0.1), 'Drug Reaction')
AIDS = ctrl.Consequent(np.arange(0, 1.1, 0.1), 'AIDS')
Diabetes = ctrl.Consequent(np.arange(0, 1.1, 0.1), 'Diabetes')
```

```python
itching['low'] = fuzz.trimf(itching.universe, [0, 0, 0.5])
itching['medium'] = fuzz.trimf(itching.universe, [0, 0.5, 1])
itching['high'] = fuzz.trimf(itching.universe, [0.5, 1, 1])

skin_rash['low'] = fuzz.trimf(skin_rash.universe, [0, 0, 0.5])
skin_rash['medium'] = fuzz.trimf(skin_rash.universe, [0, 0.5, 1])
skin_rash['high'] = fuzz.trimf(skin_rash.universe, [0.5, 1, 1])

shivering['low'] = fuzz.trimf(shivering.universe, [0, 0, 0.5])
shivering['medium'] = fuzz.trimf(shivering.universe, [0, 0.5, 1])
shivering['high'] = fuzz.trimf(shivering.universe, [0.5, 1, 1])

nodal_skin_eruptions['low'] = fuzz.trimf(nodal_skin_eruptions.universe, [0, 0, 0.5])
nodal_skin_eruptions['medium'] = fuzz.trimf(nodal_skin_eruptions.universe, [0, 0.5, 1])
nodal_skin_eruptions['high'] = fuzz.trimf(nodal_skin_eruptions.universe, [0.5, 1, 1])

continuous_sneezing['low'] = fuzz.trimf(continuous_sneezing.universe, [0, 0, 0.5])
continuous_sneezing['medium'] = fuzz.trimf(continuous_sneezing.universe, [0, 0.5, 1])
continuous_sneezing['high'] = fuzz.trimf(continuous_sneezing.universe, [0.5, 1, 1])

chills['low'] = fuzz.trimf(chills.universe, [0, 0, 0.5])
chills['medium'] = fuzz.trimf(chills.universe, [0, 0.5, 1])
chills['high'] = fuzz.trimf(chills.universe, [0.5, 1, 1])

joint_pain['low'] = fuzz.trimf(joint_pain.universe, [0, 0, 0.5])
joint_pain['medium'] = fuzz.trimf(joint_pain.universe, [0, 0.5, 1])
joint_pain['high'] = fuzz.trimf(joint_pain.universe, [0.5, 1, 1])

stomach_pain['low'] = fuzz.trimf(stomach_pain.universe, [0, 0, 0.5])
stomach_pain['medium'] = fuzz.trimf(stomach_pain.universe, [0, 0.5, 1])
stomach_pain['high'] = fuzz.trimf(stomach_pain.universe, [0.5, 1, 1])

acidity['low'] = fuzz.trimf(acidity.universe, [0, 0, 0.5])
acidity['medium'] = fuzz.trimf(acidity.universe, [0, 0.5, 1])
acidity['high'] = fuzz.trimf(acidity.universe, [0.5, 1, 1])

ulcers_on_tongue['low'] = fuzz.trimf(ulcers_on_tongue.universe, [0, 0, 0.5])
ulcers_on_tongue['medium'] = fuzz.trimf(ulcers_on_tongue.universe, [0, 0.5, 1])
ulcers_on_tongue['high'] = fuzz.trimf(ulcers_on_tongue.universe, [0.5, 1, 1])
```

```python
Fungal_infection['low'] = fuzz.trimf(Fungal_infection.universe, [0, 0, 0.5])
Fungal_infection['medium'] = fuzz.trimf(Fungal_infection.universe, [0, 0.5, 1])
Fungal_infection['high'] = fuzz.trimf(Fungal_infection.universe, [0.5, 1, 1])

Allergy['low'] = fuzz.trimf(Allergy.universe, [0, 0, 0.5])
Allergy['medium'] = fuzz.trimf(Allergy.universe, [0, 0.5, 1])
Allergy['high'] = fuzz.trimf(Allergy.universe, [0.5, 1, 1])

Drug_Reaction['low'] = fuzz.trimf(Drug_Reaction.universe, [0, 0, 0.5])
Drug_Reaction['medium'] = fuzz.trimf(Drug_Reaction.universe, [0, 0.5, 1])
Drug_Reaction['high'] = fuzz.trimf(Drug_Reaction.universe, [0.5, 1, 1])

AIDS['low'] = fuzz.trimf(AIDS.universe, [0, 0, 0.5])
AIDS['medium'] = fuzz.trimf(AIDS.universe, [0, 0.5, 1])
AIDS['high'] = fuzz.trimf(AIDS.universe, [0.5, 1, 1])

Diabetes['low'] = fuzz.trimf(Diabetes.universe, [0, 0, 0.5])
Diabetes['medium'] = fuzz.trimf(Diabetes.universe, [0, 0.5, 1])
Diabetes['high'] = fuzz.trimf(Diabetes.universe, [0.5, 1, 1])
```

```python
rule1 = ctrl.Rule(itching['high'] | skin_rash['high'] | shivering['low'] | nodal_skin_eruptions['high'] | continuous_sneezing['low'] | chills['low'] | joint_pain['low'] | stomach_pain['low'] | acidity['low'] | ulcers_on_tongue['low'], Fungal_infection['high'])
rule2 = ctrl.Rule(itching['low'] | skin_rash['low'] | shivering['high'] | nodal_skin_eruptions['low'] | continuous_sneezing['high'] | chills['high'] | joint_pain['low'] | stomach_pain['low'] | acidity['low'] | ulcers_on_tongue['low'], Allergy['high'])
rule3 = ctrl.Rule(itching['low'] | skin_rash['low'] | shivering['low'] | nodal_skin_eruptions['low'] | continuous_sneezing['low'] | chills['low'] | joint_pain['low'] | stomach_pain['high'] | acidity['high'] | ulcers_on_tongue['high'], Drug_Reaction['high'])
rule4 = ctrl.Rule(itching['high'] | skin_rash['low'] | shivering['low'] | nodal_skin_eruptions['low'] | continuous_sneezing['low'] | chills['low'] | joint_pain['low'] | stomach_pain['low'] | acidity['low'] | ulcers_on_tongue['low'], AIDS['high'])
rule5 = ctrl.Rule(itching['high'] | skin_rash['high'] | shivering['low'] | nodal_skin_eruptions['low'] | continuous_sneezing['low'] | chills['low'] | joint_pain['low'] | stomach_pain['high'] | acidity['low'] | ulcers_on_tongue['low'], Diabetes['high'])
```

```python
disease_ctrl_1 = ctrl.ControlSystem([rule1])
disease_ctrl_2 = ctrl.ControlSystem([rule2])
disease_ctrl_3 = ctrl.ControlSystem([rule3])
disease_ctrl_4 = ctrl.ControlSystem([rule4])
disease_ctrl_5 = ctrl.ControlSystem([rule5])

disease_diag_1 = ctrl.ControlSystemSimulation(disease_ctrl_1)
disease_diag_2 = ctrl.ControlSystemSimulation(disease_ctrl_2)
disease_diag_3 = ctrl.ControlSystemSimulation(disease_ctrl_3)
disease_diag_4 = ctrl.ControlSystemSimulation(disease_ctrl_4)
disease_diag_5 = ctrl.ControlSystemSimulation(disease_ctrl_5)
```

```python
symptom_values = [1, 1, 10, 1, 10, 10, 1, 1, 1, 1]
for i, sym in enumerate([itching, skin_rash, shivering, nodal_skin_eruptions, continuous_sneezing, chills, joint_pain, stomach_pain, acidity, ulcers_on_tongue]):
  disease_diag_1.input[sym.label] = symptom_values[i]
  disease_diag_2.input[sym.label] = symptom_values[i]
  disease_diag_3.input[sym.label] = symptom_values[i]
  disease_diag_4.input[sym.label] = symptom_values[i]
  disease_diag_5.input[sym.label] = symptom_values[i]
```

```python
disease_diag_1.compute()
```

```python
print("Likelihood of having the disease:", disease_diag_1.output['Fungal infection'])
```
